```python
from __future__ import absolute_import, division, print_function, unicode_literals


import tensorflow as tf

import pandas as pd
```

```python
CSV_COLUMN_NAMES = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
SPECIES = ['Setosa', 'Versicolor', 'Virginica']
# Lets define some constants to help us later on
```

```python
train_path = tf.keras.utils.get_file(
    "iris_training.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_training.csv")
test_path = tf.keras.utils.get_file(
    "iris_test.csv", "https://storage.googleapis.com/download.tensorflow.org/data/iris_test.csv")

train = pd.read_csv(train_path, names=CSV_COLUMN_NAMES, header=0)
test = pd.read_csv(test_path, names=CSV_COLUMN_NAMES, header=0)
# Here we use keras (a module inside of TensorFlow) to grab our datasets and read them into a pandas dataframe
```

```python
train.head()
```

```python
train_y = train.pop('Species')
test_y = test.pop('Species')
train.head() # the species column is now gone
```

```python
train.shape  # we have 120 entires with 4 features
```

```python
def input_fn(features, labels, training=True, batch_size=256):
    # Convert the inputs to a Dataset.
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))

    # Shuffle and repeat if you are in training mode.
    if training:
        dataset = dataset.shuffle(1000).repeat()

    return dataset.batch(batch_size)
```

```python
# Feature columns describe how to use the input.
my_feature_columns = []
for key in train.keys():
    my_feature_columns.append(tf.feature_column.numeric_column(key=key))
print(my_feature_columns)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(30, activation='relu', input_shape=(train.shape[1],)),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(train.values, train_y.values, epochs=50, batch_size=32)
model.evaluate(test.values, test_y.values)
```

```python
import numpy as np

features = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
predict_input = []

print("Please type numeric values as prompted. Press Enter when done.")

for feature in features:
    while True:
        val = input(f"{feature}: ").strip()
        if val == "":
            print("Input cannot be empty. Please enter a number.")
            continue
        try:
            val_float = float(val)
            predict_input.append(val_float)
            break
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

# Convert to 2D array for Keras
predict_array = np.array([predict_input])

# Make prediction
pred_probs = model.predict(predict_array)
class_id = np.argmax(pred_probs[0])
probability = pred_probs[0][class_id]

print(f'Prediction is "{SPECIES[class_id]}" ({probability*100:.1f}%)')
4
```

```python
import numpy as np

# Example input
predict_x = {
    'SepalLength': [5.1, 5.9, 6.9],
    'SepalWidth': [3.3, 3.0, 3.1],
    'PetalLength': [1.7, 4.2, 5.4],
    'PetalWidth': [0.5, 1.5, 2.1],
}

# Convert dict to 2D array for Keras
predict_array = np.column_stack([
    predict_x['SepalLength'],
    predict_x['SepalWidth'],
    predict_x['PetalLength'],
    predict_x['PetalWidth']
])

# Make predictions
pred_probs = model.predict(predict_array)

# Print results
for i, probs in enumerate(pred_probs):
    class_id = np.argmax(probs)
    probability = probs[class_id]
    print(f'Sample {i+1} prediction is "{SPECIES[class_id]}" ({probability*100:.1f}%)')
```
